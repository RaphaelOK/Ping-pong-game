```python
import pygame
import time

# Set up the game window
pygame.init()
width = 800
height = 600
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Pong Game")

# Set up the game objects
ball = pygame.Rect(width // 2 - 10, height // 2 - 10, 20, 20)
player1 = pygame.Rect(50, height // 2 - 50, 20, 100)
player2 = pygame.Rect(width - 70, height // 2 - 50, 20, 100)

# Set up the game variables
ball_speed_x = 5
ball_speed_y = 5
player_speed = 7
score1 = 0
score2 = 0
font = pygame.font.Font(None, 50)

# Game loop
running = True
while running:
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    # Move the ball
    ball.x += ball_speed_x
    ball.y += ball_speed_y
    
    # Check for collisions with walls and paddles
    if ball.top <= 0 or ball.bottom >= height:
        ball_speed_y *= -1
    if ball.left <= 0:
        ball_speed_x *= -1
        score2 += 1
        ball.center = (width // 2, height // 2)
    if ball.right >= width:
        ball_speed_x *= -1
        score1 += 1
        ball.center = (width // 2, height // 2)
    if ball.colliderect(player1) or ball.colliderect(player2):
        ball_speed_x *= -1
    
    # Move the players
    keys = pygame.key.get_pressed()
    if keys[pygame.K_w] and player1.top > 0:
        player1.y -= player_speed
    if keys[pygame.K_s] and player1.bottom < height:
        player1.y += player_speed
    if keys[pygame.K_UP] and player2.top > 0:
        player2.y -= player_speed
    if keys[pygame.K_DOWN] and player2.bottom < height:
        player2.y += player_speed
    
    # Clear the screen
    screen.fill((0, 0, 0))
    
    # Draw the game objects
    pygame.draw.rect(screen, (255, 255, 255), player1)
    pygame.draw.rect(screen, (255, 255, 255), player2)
    pygame.draw.ellipse(screen, (255, 255, 255), ball)
    pygame.draw.aaline(screen, (255, 255, 255), (width // 2, 0), (width // 2, height))
    
    # Draw the scores
    score_text1 = font.render(str(score1), True, (255, 255, 255))
    score_text2 = font.render(str(score2), True, (255, 255, 255))
    screen.blit(score_text1, (width // 2 - 70, 50))
    screen.blit(score_text2, (width // 2 + 50, 50))
    
    # Update the display
    pygame.display.update()
    
    # Add delay to the game loop
    time.sleep(0.01)

# Quit the game
pygame.quit()
```
